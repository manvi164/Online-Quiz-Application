```python
# ---------- Database Connection ----------
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="1234",
    database="inventory_db"
)

cursor = db.cursor() 

# ---------- Product Class ----------
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def availability(self):
        if self.stock > 0:
            return "Available"
        else:
            return "Out of Stock"

# ---------- Add Product ----------
def add_product():
    name = input("Product name: ")
    price = int(input("Price: "))
    stock = int(input("Stock: "))

    p = Product(name, price, stock)

    query = "INSERT INTO products (name, price, stock) VALUES (%s, %s, %s)"
    cursor.execute(query, (p.name, p.price, p.stock))
    db.commit()

    print("Product Added |", p.availability())

# ---------- View Products ----------
def view_products():
    cursor.execute("SELECT * FROM products")
    data = cursor.fetchall()

    for row in data:
        print(row)

# ---------- Main Menu ----------
while True:
    print("\n1. Add Product")
    print("2. View Products")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_product()
    elif choice == "2":
        view_products()
    elif choice == "3":
        print("Thank you!")
        break
    else:
        print("Invalid choice")
```
